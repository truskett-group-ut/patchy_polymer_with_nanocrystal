```python
model_dict={}

for mod in ['fe', 'cnts', 'amps', 'efas', 'rbfs']:
    modelfile='model_'+ mod + '.json'
    json_file = open(modelfile, 'r')
    model_json = json_file.read()
    json_file.close()
    model_dict[mod] = model_from_json(model_json)
    weightsfile='model_'+ mod + '.h5'
    model_dict[mod].load_weights(weightsfile)
```

```python
r = np.arange(0.05, 7.01, 0.05)
rnorm = np.transpose(r_scaler.fit_transform( np.transpose(np.array([r]))))[0]
```

```python
functional_helper = FunctionalFitHelper()
functional_helper.num_cnts=35
```

```python
functional_helper.InitializeOldModel(model_dict['cnts'], model_dict['amps'], model_dict['efas'], model_dict['rbfs'])
loaded_model=functional_helper.BuildModel(rnorm)
```

```python
gr=loaded_model.predict(np.array([[-0.70847558, -0.44129471,  0.19837407,  0.99963276, -0.72520597]]))[0]
```
